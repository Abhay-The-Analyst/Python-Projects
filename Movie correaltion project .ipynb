```python
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv(r'C:\Users\abhay\OneDrive\Desktop\movies.csv')
df.head()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df = df.reset_index(drop = True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
df = df.sort_values(by = ['gross'] , inplace = False , ascending   = False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title("Budget vs Gross", fontsize=16, color='Red')
plt.xlabel("Budget for film")
plt.ylabel("Gross Earning")
plt.ticklabel_format(style='plain', axis='both')

plt.show()
```

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws = {"color": "Red"} , line_kws = {"color" : "blue"})
plt.title("Budget vs Gross", fontsize=16)
plt.xlabel("Budget for Film")
plt.ylabel("Gross Earnings")
plt.ticklabel_format(style='plain', axis='both')
plt.show()
```

```python
df.corr(numeric_only=True , method = 'pearson')
```

```python
correlaton_matrix = df.corr(numeric_only=True , method = 'pearson') 
```

```python
sns.heatmap(correlaton_matrix, annot = True)
plt.title("Correlation Matric for numeric", fontsize=16)
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
df.head()
```

```python
df_numerizer = df
for col_name in df_numerizer.columns:
    if (df_numerizer[col_name].dtype =="object"):
        df_numerizer[col_name] = df_numerizer[col_name].astype("category")
        df_numerizer[col_name] = df_numerizer[col_name].cat.codes
df_numerizer.head()
        
```

```python
df.head()
```

```python
correlaton_matrix = df_numerizer.corr(numeric_only=True , method = 'pearson') 
sns.heatmap(correlaton_matrix, annot = True)
plt.title("Correlation Matric for numeric", fontsize=16)
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```
