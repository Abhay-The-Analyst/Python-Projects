```python
mental_health_by_arrangement = df.groupby('Work_Arrangement')['Mental_Health_Status'].value_counts().unstack().fillna(0)
mental_health_by_arrangement 
```

```python
mental_health_by_arrangement.plot(kind='bar',  figsize=(10, 6), colormap='Set3')
plt.title("Mental Health Status by Work Arrangement")
plt.xlabel("Work Arrangement")
plt.ylabel("Number of Respondents")
plt.legend(title="Mental Health Status", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
bins = [20, 30, 40, 50, 60, 70]
labels = ['21-30', '31-40', '41-50', '51-60', '61-70']
df['Age_Group'] = pd.cut(df['Age'], bins=bins ,   labels=labels, right=False)
```

```python
df.head()
```

```python
wlb_by_age_arrangement = df.groupby(['Age_Group', 'Work_Arrangement'])['Work_Life_Balance_Score'].mean().unstack()
wlb_by_age_arrangement 
```

```python
wlb_by_age_arrangement.plot(kind='bar', figsize=(10, 6), colormap='coolwarm')
plt.title('Avg Work-Life Balance Score by Age Group and Work Type')
plt.xlabel('Age Group')
plt.ylabel('Avg Work-Life Balance Score')
plt.legend(title='Work Arrangement')
plt.tight_layout()
plt.show()
```

```python
wlb_salary_range = df.groupby(["Salary_Range" , "Work_Arrangement"])["Work_Life_Balance_Score"].mean()
wlb_salary_range.plot(kind = "bar" , figsize = (10,6) , colormap ='Accent')
plt.title("Avg Work-Life Balance Score by Salary Range")
plt.xlabel("Salary Range")
plt.ylabel("Avg Balance Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
industry_wlbs = df.groupby(["Industry" , "Work_Arrangement"])["Work_Life_Balance_Score"].mean().unstack()
industry_wlbs.plot(kind = "bar", figsize = (10,6), colormap = 'viridis' )
```

```python
burnout_by_industry = df.groupby('Industry')['Burnout_Level'].value_counts().unstack().fillna(0)
burnout_by_industry.plot(kind='bar', figsize=(12, 6), colormap='coolwarm')
plt.title("Burnout Levels Across Industries")
plt.xlabel("Industry")
plt.ylabel("Number of Respondents")
plt.legend(title="Burnout Level", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
